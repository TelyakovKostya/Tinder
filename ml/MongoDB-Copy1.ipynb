```python
import datetime;
import random

person_data = {}


for i, person in enumerate(data_new['results']):    
    print(i)
    person_data = {
            'timestamp': datetime.datetime.now().timestamp(),
            'user_id': 0,
            'person': {
                '_id': data_new['results'][i]['_id'],
                'name': data_new['results'][i]['name'],
                'distance_mi': data_new['results'][i]['distance_mi'],
                'photos':[]
            }           
    }

    try:
        person_data['person']['age'] = datetime.datetime.now().year - int(data_new['results'][i]['birth_date'][:4])
    except IndexError:
        person_data['person']['age'] = None
        
    try:
        person_data['person']['schools'] =  data_new['results'][i]['schools'][-1]['name']
    except IndexError:
        person_data['person']['schools'] =  None
    
    try:
        person_data['person']['job'] = data_new['results'][i]['jobs'][-1]['company']['name']
    except KeyError:
        person_data['person']['job'] = data_new['results'][i]['jobs'][-1]['title']['name']
    except Exception:
        person_data['person']['job'] = None
        
    try:
        person_data['person']['bio'] = data_new['results'][i]['bio']
    except KeyError:
        person_data['person']['bio'] = None
        
        
        #db.train_img_set.drop()

    for j, photo in enumerate(data_new['results'][i]['photos']):
        photo_dict={}
        
        photo_dict['id'] = data_new['results'][i]['photos'][j]['id'] 
        try:
            photo_dict['link'] = data_new['results'][i]['photos'][j]['url']                        
        except Exception:
            print('–û—à–∏–±–∫–∞ –≤ –ª–∏–Ω–∫–µ photo')
            photo_dict['link'] = None                       

        try:
            photo_dict['model_score'] = [random.uniform(0,1),random.uniform(0,1)]
                        
        except Exception:
            print('–û—à–∏–±–∫–∞ –≤ score')
            photo_dict['model_score'] = None                        
        
        #print(photo_dict)
        person_data['person']['photos'].append(photo_dict)                         
    
        
        
    try:
        person_data['person']['model_result'] = random.choice(['like', 'pass'])
    except Exception:
        person_data['person']['model_result'] = None
            
  
    print(person_data)
    db.recommendation_model.save(person_data)

```
